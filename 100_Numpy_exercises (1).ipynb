#### 29. How to round away from zero a float array ? (★☆☆)

````python
import numpy as np

def round_away_from_zero(x):
  return np.where(x>0, np.ceil(x), np.floor(x))

X = np.array([1.2,1.8, -1.2, -1.8, 0.0])
print(round_away_from_zero(X))
````

#### 35. How to compute ((A+B)*(-A/2)) in place (without copy)? (★★☆)

#### 48. Print the minimum and maximum representable values for each numpy scalar type (★★☆)

````python
for dtype in [np.int8, np.int32, np.int64, np.uint8, np.uint32, np.uint64, np.float32, np.float64]:

  if np.issubdtype(dtype, np.integer):
    info = np.iinfo(dtype)
  else:
    info = np.finfo(dtype)

  print(dtype.__name__, info.min, info.max)

````

#### 62. Considering two arrays with shape (1,3) and (3,1), how to compute their sum using an iterator? (★★☆)

````python
a = np.array([[1, 2, 3]])   # shape (1,3)
b = np.array([[10], [20], [30]])  # shape (3,1)

result = np.empty((3, 3), dtype=int)

for (i, j), val in np.ndenumerate(np.zeros((3, 3))):
    result[i, j] = a[0, j] + b[i, 0]

print(result)

````
